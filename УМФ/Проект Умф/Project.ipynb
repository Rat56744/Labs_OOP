#### Задача:  
Струна конечной длины. Левый конец струны закреплен, правый движется по закону:  
$$x \left(t \right) =  \alpha(1 - e^{\frac{-t}{\tau} } )$$
Параметры: плотность струны, натяжение струны, длина струны, амплитуда импульса a и характерное время нарастания/спадания импульса τ. 

### 1. Постановка задачи

Рассмотрим физическую модель конечной струны длины L, один конец закреплен, а правый движется по закону:
 $$x \left(t \right) =  \alpha(1 - e^{\frac{-t}{\tau} } )$$ 
, где $\alpha$ - амплитуда импульса струны, а τ - характерное время нарастания/спадания импульса.

Сопротивлением воздуха и массой струны можно пренебречь. Струна считается абсолютно тонкой, однородной и нерастяжимой. Свободное колебание струны на правом конце можно описать граничным условием второго рода, закрепленный конец струны подчинаяется граничному условию первого рода.

**Граничные условия:** 
1. Граничное условие для левого конца струны $\left(x = 0 \right)$: $U \left( t,0 \right) = l \left( t \right)$.
2. Граничное условие для правого конца струны $\left(x = L \right)$: $U \left( t,L \right) = r \left( t \right)$.

**Начальные условия:**  
1. Начальное смещение струны: $U \left( 0,x \right) = f \left( x \right)$, где функция $f \left( x \right)$ определяет начальное положение струны в момент времени $t = 0$.
2. Начальная скорость струны: $\frac{\partial U}{\partial t} = g \left( x \right)$, где функция $g \left( x \right)$ описывает начальную скорость струны в момент времени $t = 0$.

**Обоснование условий:**  
1. Начальные условия: Они задают начальное состояние струны и определяют, как она выглядит в момент времени $t = 0$. Если струна покоится, $f \left( x \right) = 0$ и $g \left( x \right) = 0$.
2. Граничные условия: Они указывают на то, что левый конец струны закреплён и неподвижен, а правый движется в соответствии с законом $l \left( t \right) = 0$, $x\left( t \right) =  \alpha(1 - e^{\frac{-t}{\tau} } )$.

Дифференциальное уравнение колебаний тогда можно записать как:
$$
U_{tt} = 10202 \cdot U_{xx} \\
\begin{cases}
U(x,0) = 0\\
U_t (x,0) = 0\\
U(0,t) = 0\\
U_x(L,t) = \alpha(1 - e^{\frac{-t}{\tau} }  )
\end{cases}
$$

- $U(x, t)$: смещение струны в точке x и в момент времени t
- $L$ - длина струны ($0.1м$)
- $\alpha$ - амплитуда импульса $[0.005,0.01]м$
- $t_{0}$- момент времени, в который начинается действие импульса $(0.001 сек)$  
- $\tau$ - характерное время нарастания/спадания импульса $[0.025,0.05]сек$  
$a^2$ = $\frac{N}{\rho S} ≈ 10202$,  
$N$ -  натяжение  струны ($40Н$),     
$\rho$ - плотность материала струны ($7800 \;кг/м^3$),   
$S$ - площадь сечения струны ($5.0265\cdot 10^{-7}\;м^2$)  
Возьмем струну диаметром $0.8 мм$, тогда площадь её сечения будет равна: $S = \frac{d^2}{4}\pi = 5.0265\cdot 10^{-7}$

### 2. Теория

**Термины:**  
- **Сетка** - Дискретное представление пространства, состоящее из узлов (точек), в которых задаются значения функции или ее производных.  
    - **Равномерная сеточная область** - Сетка, в которой расстояния между соседними узлами одинаковы на всей области. 
    - **Неравномерная сеточная область** - Сетка, в которой расстояния между узлами могут варьироваться, позволяя более точно моделировать функции с высоким градиентом или другими особенностями. 
    - **Явный точечный шаблон** - Использует известные значения сетки.
    - **Неявный точечный шаблон** - Использует неизвестные значения, которые находятся с помощью уравнения.
* **Сеточная область** - Края области, в которой решается задача; отличается от сетки тем, что это не точки (узлы), а контуры или линии, определяющие пределы вычисляемой области.   
    - **Граница сеточной области** - Края области, в которой решается задача; отличается от сетки тем,  что это не точки (узлы), а контуры или линии, определяющи пределы вычисляемой области.  
    - **Область определения решения** - Сеточная область — это конкретная область, в которой решается численная задача, в то время как область определения решени 
может быть более широкой, включая всю функцию.  
*  **Сеточная функция** - Функция, определенная только на узлах сетки, не обязательно обеспечивающая непрерывность между узлами.   

*  **Аналитическая функция** - Функция, заданная на всей области определения и обладающая свойствами гладкости и непрерывности.    
 
*  **Сходимость решения** - Свойство численного метода, при котором решение, получаемое с помощью сеток, стремится к настоящему (точному) решению с уменьшением размера шага сетки.        

* **Точность решения** - Степень, с которой численное решение приближается к точному решению проблемы; часто зависит от порядка метода и расстояния между узлами.  

* **Сеточная задача** - Задача, в которой уравнения (и/или условия) решаются на дискретной сетке с заданными краевыми и начальными условиями для получения численного решения.

### 3. Постановка краевой задачи

**Обезразмеривание физических величин:**  
Идея обезразмеривания заключается в переходе от абсолютных значений к относительным.  
  
Квадратная, равномерная сетка на отрезке $x \in [0,1]$ с шагом $h$ и $t \in [0,T]$ с шагом $p$:  
$$x_k = h \cdot k, \;\;\; k = 0,1,..., N, \;\;\; h\cdot N = 1$$
$$t_p = h \cdot p, \;\;\; j = 0,1,..., P, \;\;\; j\cdot P = 1$$  

Обезразмерим наши переменные:

$$ x \in [0,L] \;\; => \;\; \overline{x} \in [0,1], \;\;\; x  \; => \;  \overline{x} L$$
$$ t \in [0,T] \;\; => \;\; \overline{t} \in [0,1], \;\;\; t  \; => \;  \overline{t} T$$    
  
$L = 0.1м$,  
$T = 1 сек$.


**Производные:**  
Перепишу производные функции $U(\overline{x},\overline{t})$ по $t$ и $x$ согласно правилу взятия производной от сложной функции:
$$U_t=U_{\overline{t}} \cdot \overline{t}_{t}= \frac{1}{T} \cdot U_{\overline{t}}$$
$$U_{tt}= \frac{1}{T^2} \cdot U_{\overline{t} \overline{t}}$$
$$U_x=U_{\overline{x}} \cdot \overline{x}_{x}= \frac{1}{L} \cdot U_{\overline{x}}$$
$$U_{xx}= \frac{1}{L^2} \cdot U_{\overline{x} \overline{x}}$$

Краевая задача после замены:
$$\frac{1}{T^2} \cdot U_{\overline{t}\overline{t}}= \frac{1}{L^2}10202\cdot U_{\overline{x}\overline{x}}$$
$$\begin{cases}
U(\overline{x},0)=0 \\
\frac{1}{T} U_{\overline t}(\overline x,0) = 0 \\
U(0,\overline t) = 0 \\
U(1,\overline t)= \alpha(1 - e^{\frac{-t}{\tau T}})  \\
\end{cases}$$
