## 2. Теоретические выкладки

**Термины:**  
* **Сетка** - это счетная совокупность точек (так называемых узлов сетки) алгебраического пространств.  
    $\omega = (x_0,x_1, ... , x_n)$
    - **Равномерная сетка** - сетка, в которой расстояния между соседними узлами одинаковы на всей области.  
    - **Неравномерная сетка** - сетка, в которой расстояния между узлами могут варьироваться, позволяя более точно моделировать функции с высоким градиентом  или другими особенностями.  
* **Сеточная область** - часть пространства, в которой расположены узлы сетки и где
происходит решение задачи, ограниченная краевыми или граничными условиями.  
    * **Граница сеточной области** - края области, в которой решается задача; отличается от сетки тем,  что это не точки (узлы), а контуры или линии, определяющи пределы вычисляемой области.  
    * **Область определения решения** - сеточная область — это конкретная область, в которой решается численная задача, в то время как область определения решени
может быть более широкой, включая всю функцию.  
*  **Сеточная функция** - функция, определенная только на узлах сетки, не обязательно обеспечивающая непрерывность между узлами    ($y_k = y(x_k))$.  
*  **Аналитическая функция** - функция, заданная на всей области определения и обладающая свойствами гладкости и непрерывности.  
*  **Сходимость решения** - свойство численного метода, при котором решение, получаемое с помощью сеток, стремится к настоящему (точному) решению уменьшением размера шага сетки.  
* **Точность решения** - степень, с которой численное решение приближается к точному
решению проблемы; часто зависит от порядка метода и расстояни между узлами.  
* **Краевая задача** - задача с условиями, задающими значения или свойства
решения на границах области.  
* **Сеточная задача** - задача, в которой уравнения (и/или условия) решаются на
дискретной сетке с заданными краевыми и начальными условиям для получения численного решения.

## 4. Постановка КЗ

### Постановка краевой задачи для колебаний струны

В данной лабораторной работе мы будем исследовать колебания струны, используя подход к решению дифференциального уравнения для одномерной волны. Важно правильно сформулировать краевую задачу (КЗ) и убедиться, что все переменные обезразмерены.

#### Обезразмеривание переменных

Для удобства расчета и соблюдения условий устойчивости мы проведем обезразмеривание переменных следующим образом:

1. **Временная переменная:**
  $
   t = n \cdot \delta t \quad (n = 1 \ldots N, \quad \delta t = \frac{T}{N})
   $ \
   при этом $T$ — время, принимаем равным 1 секунде.

2. **Пространственная переменная:**
   $
   x = m \cdot \delta x \quad (m = 1 \ldots M, \quad \delta x = \frac{L}{M})
   $ \
   где $L$ — длина струны, равная 10 см (или 0.1 метра).
   
Перепишу производные функции $U(\overline{x},\overline{t})$ по $t$ и $x$ согласно правилу взятия производной от сложной функции:
$$U_t=U_{\overline{t}} \cdot \overline{t}_{t}= \frac{1}{T} \cdot U_{\overline{t}}$$
$$U_{tt}= \frac{1}{T^2} \cdot U_{\overline{t} \overline{t}}$$
$$U_x=U_{\overline{x}} \cdot \overline{x}_{x}= \frac{1}{L} \cdot U_{\overline{x}}$$
$$U_{xx}= \frac{1}{L^2} \cdot U_{\overline{x} \overline{x}}$$

#### Исходная краевая задача

Наша краевая задача будет выглядеть как:


$
U_{tt} = 9921.3471 U_{xx} + \frac{a}{0.006} \sin(kx)
$


С учетом выбранных параметров, модель становится:

$
\frac{1}{T^2}U_{\overline{tt}} = 9921.3471\cdot\frac{1}{L^2}U_{\overline{xx}}+ \frac{a}{0.006}sin(k\overline{x})
$

и определяем начальные и краевые условия:

1. **Начальное условие 1:**
   $
   U(x, 0) = 0
   $

2. **Начальное условие 2:**
   $
   U_t(x, 0) = 0
   $

3. **Краевое условие 1 :**
   $
   U(0, t) = 0
   $

4. **Краевое условие 2 :**
   $
   U(1, t) = 0
   $

### Дополнительный вопрос

Решение продолжает оставаться правильным после одной секунды благодаря свойству линейности уравнения и сохранению формы волнового процесса. Колебания последовательны и периодичны, поэтому, даже если мы рассматриваем только один временной шаг, физические свойства системы остаются неизменными, а волна будет продолжать распространяться, сохраняя свою форму.

$$p=0,1,...,P$$
Аппроксимация начальных условий:
$$U_m^0=0 \; \; \; \; (2.4)$$
$$\frac{U_m^1-U_m^0}{\tau} = \frac{\tau}{2 \!} \cdot \frac{a}{0.006}sin(kx) + o(\tau^2) \; \; \; \; (2.5)$$
$$m=0,1,...,M$$

Во внутренних узлах уравнения $(2.1), (2.2), (2.3), (2.4)$ и $(2.5)$ аппроксимируют исходное дифференциальное уравнение со вторым порядком точности $o(\tau^2 + h^2)$.

Схема является абсолютно устойчивой.

Из уравненя $(2.1)$ выражу значения сеточной функции $U_m^p$ в узлах $(p+1)$:
$$-\frac{9921.3471}{L^2h^2} \cdot U_{m-1}^{p+1} + \left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{9921.3471}{L^2h^2} \right) \cdot U_m^{p+1} - \frac{9921.3471}{L^2h^2} \cdot U_{m+1}^{p+1} = \frac{a}{0.006}sin(kx) \cdot (U_m^{p-1} - U_m^p) \; \; \; \; (2.6)$$
Составлю матрицу для решения дифференциального уравнения методом прогонки:
$$A_i = -\frac{9921.3471}{L^2h^2}; \; \; B_i = -\frac{9921.3471}{L^2h^2}; \; \; -C_i = \left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{9921.3471}{L^2h^2} \right); \; \; \; \; (2.7)$$
$$-F_i = \frac{a}{0.006}sin(kx) - \frac{1}{T^2 \tau^2} \cdot (U_m^{p-1} - U_m^p)$$
Из краевых условий:
$$U_0^p = 0 \; \; \implies \; \; \chi_1 = \mu_1 = 0$$
$$U_M^p = 0 \; \; \implies \; \; \chi_2 = \mu_2 = 0$$
Тогда матрица будет иметь вид:
$$ A = \left( \begin{matrix}
1 & 0 & ... & ... & ... & ... & ... \\
-\frac{9921.3471}{L^2h^2} & \left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{9921.3471}{L^2h^2} \right) & -\frac{9921.3471}{L^2h^2} & 0 & ... & ... & ... \\
0 & -\frac{9921.3471}{L^2h^2} & \left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{9921.3471}{L^2h^2} \right) & -\frac{0.0078}{L^2h^2} & 0 & ... & ... \\
... & 0 & -\frac{9921.3471}{L^2h^2} & \left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{9921.3471}{L^2h^2} \right) & -\frac{9921.3471}{L^2h^2} & 0 & ... \\
... & ... & ... & ... & ... & ... & ... \\
... & ... & 0 & -\frac{9921.3471}{L^2h^2} & \left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{9921.3471}{L^2h^2} \right) & -\frac{9921.3471}{L^2h^2} \\
... & ... & ... & ... & ... & 0 & 1
\end{matrix} \right); \; \; AU = F \; \; \; \; (2.8)$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Шаг 1: Определение параметров
L = 1.0  # длина области
T = 1.0   # время
M = 1000   # количество пространственных узлов
P = 1000    # количество временных шагов
tau = 0.01  # временной шаг
h = L / M  # пространственный шаг

# Константы
a = 1.0  # коэффициент для синуса
k = 1 * np.pi / L  # волновое число

# Шаг 2: Инициализация матрицы и векторов
A = np.zeros((M-1, M-1))  # Заполняем нулями
F = np.zeros(M-1)  # Вектор правой части

# Заполнение матрицы A и вектора F
for m in range(1, M-1):
    A[m-1, m-1] = (1 / (tau**2)) + 2 * (a / (h**2))
    if m > 1:
        A[m-1, m-2] = - (a / (h**2))  # A_i
    if m < M - 2:
        A[m-1, m] = - (a / (h**2))  # B_i

# Заполнение вектора F
for m in range(1, M-1):
    F[m-1] = (a / 0.006) * np.sin(k * m * h)

# Шаг 3: Начальные условия
U = np.zeros((M, P))  # Решение инициализировано нулями
U[:, 0] = 0  # Начальные условия U[0, :] = 0

# Шаг 4: Применение метода прогонки
for p in range(1, P):
    # Прогонка
    # Прямой ход
    c = np.zeros(M-1)
    b = np.zeros(M-1)

    c[0] = -A[0, 1] / A[0, 0]
    U[1, p] = F[0] / A[0, 0]

    for m in range(1, M-2):
        b[m] = A[m, m] + A[m, m-1] * c[m-1]
        c[m] = -A[m, m+1] / b[m]
        U[m+1, p] = (F[m] - A[m, m-1] * U[m-1, p]) / b[m]

    # Обратный ход
    for m in range(M-3, -1, -1):
        U[m+1, p] += c[m] * U[m+2, p]

# # Шаг 5: Визуализация результатов
x = np.linspace(0, L, M)
for p in range(0, P, 5):
     plt.plot(x, U[:, p], label=f't={p*tau:.2f} s')

plt.title('Решение неявного разностного уравнения')
plt.xlabel('x')
plt.ylabel('U(x, t)')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Параметры задачи
L = 1.0           # Длина струны (м)
T = 10.0          # Натяжение струны (Н)
rho = 0.01        # Плотность струны (кг/м)
a_force = 1.0     # Амплитуда силы (Н)
k = 2 * np.pi     # Волновое число (рад/м)

N = 7000           # Количество временных шагов
M = 100            # Количество пространственных узлов
delta_t = 0.001   # Шаг по времени (с)
h = L / M         # Шаг по пространству (м)

c = np.sqrt(T / rho)           # Скорость волны
s = (c * delta_t / h) ** 2     # Коэффициент схемы


# Сетка и массивы для решения
U_nm = np.zeros((N + 1, M + 1))  # Решение U(n, m)
alpha_m = np.zeros(M + 1)        # Прогоночные коэффициенты α
beta_m = np.zeros(M + 1)         # Прогоночные коэффициенты β

# Граничные условия
U_nm[:, 0] = 0      # U(0, t) = 0
U_nm[:, M] = 0      # U(L, t) = 0

# Вычисление периода волны
wave_period = 2 * np.pi / (k * c)

# Уменьшение времени моделирования
N = int(wave_period / delta_t)  # Количество временных шагов на период
U_nm = np.zeros((N + 1, M + 1))  # Обновление массива решения

# Начальные условия
U_nm[0, :] = 0  # U(x, 0) = 0
U_nm[1, :] = 0  # U_t(x, 0) = 0

# Функция неоднородной силы
def force_term(x, t):
    return (a_force * np.sin(k * x)) * (delta_t ** 2 / rho)

# Неявный метод + метод прогонки
for n in range(1, N):
    # Правая часть схемы (влияние предыдущих временных слоёв)
    f_m = np.zeros(M - 1)
    for m in range(1, M):
        f_m[m - 1] = (2 * U_nm[n, m] - U_nm[n - 1, m]) + force_term(m * h, n * delta_t)

    # Коэффициенты прогонки
    A = -s
    B = -s
    C = 2 + 2 * s

    # Прямой ход прогонки
    alpha_m[1] = 0
    beta_m[1] = 0
    for m in range(1, M):
        denom = C - A * alpha_m[m]
        alpha_m[m + 1] = B / denom
        beta_m[m + 1] = (f_m[m - 1] - A * beta_m[m]) / denom

    # Обратный ход прогонки
    for m in range(M - 1, 0, -1):
        U_nm[n + 1, m] = alpha_m[m + 1] * U_nm[n + 1, m + 1] + beta_m[m + 1]

# Параметры задачи
L = 1.0         # Длина струны (м)
T = 10.0        # Натяжение струны (Н)
rho = 0.01      # Плотность струны (кг/м)
a = 1.0         # Амплитуда силы (Н)
k = 3 * np.pi   # Волновое число (рад/м)

Nx = 100  # Число узлов по пространству
Nt = 1000  # Число временных шагов
dx = L / (Nx - 1)
dt = 0.0001  # Временной шаг

# Коэффициент разностной схемы
c = np.sqrt(T / rho)  # Скорость волны
s = (c * dt / dx) ** 2


# Инициализация сетки
x = np.linspace(0, L, Nx)
U = np.zeros((Nt, Nx))

# Начальные условия (струна покоится)
U[0, :] = 0
U[1, :] = 0

# Расчет на основе разностной схемы "Крест"
for n in range(1, Nt - 1):
    for i in range(1, Nx - 1):
        U[n + 1, i] = (2 * U[n, i] - U[n - 1, i] +
                       s * (U[n, i + 1] - 2 * U[n, i] + U[n, i - 1]) +
                       (a * dt ** 2 / rho) * np.sin(k * x[i]))

# Визуализация результата
X, T_grid = np.meshgrid(x, np.linspace(0, Nt * dt, Nt))

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, T_grid, U, cmap='viridis')
ax.set_title("Колебания струны под воздействием неоднородной силы")
ax.set_xlabel("x (м)")
ax.set_ylabel("t (с)")
ax.set_zlabel("U(x, t)")
plt.show()
```
