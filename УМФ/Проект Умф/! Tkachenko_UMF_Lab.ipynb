### **2. Теоретические выкладки (1 лабораторная работа)**

**Термины:**  
* **Сетка** - это счетная совокупность точек (так называемых узлов сетки) алгебраического пространств.  
    - **Равномерная сетка** - сетка, в которой расстояния между соседними узлами одинаковы на всей области.  
    - **Неравномерная сетка** - сетка, в которой расстояния между узлами могут варьироваться, позволяя более точно моделировать функции с высоким градиенто  или другими особенностями.  
* **Сеточная область** - часть пространства, в которой расположены узлы сетки и где
происходит решение задачи, ограниченная краевыми или граничными условиями.  
    - **Граница сеточной области** - края области, в которой решается задача; отличается от сетки тем,  что это не точки (узлы), а контуры или линии, определяющи пределы вычисляемой области.  
    - **Область определения решения** - сеточная область — это конкретная область, в которой решается численная задача, в то время как область определения решени 
может быть более широкой, включая всю функцию.  
*  **Сеточная функция** - функция, определенная только на узлах сетки, не обязательно обеспечивающая непрерывность между узлами.  
*  **Аналитическая функция** - функция, заданная на всей области определения и обладающая свойствами гладкости и непрерывности.  
*  **Сходимость решения** - свойство численного метода, при котором решение, получаемое с помощью сеток, стремится к настоящему (точному) решению уменьшением размера шага сетки.  
* **Точность решения** - степень, с которой численное решение приближается к точному
решению проблемы; часто зависит от порядка метода и расстояни между узлами.  
* **Краевая задача** - задача с условиями, задающими значения или свойства 
решения на границах области.  
* **Сеточная задача** - задача, в которой уравнения (и/или условия) решаются на
дискретной сетке с заданными краевыми и начальными условиям для получения численного решения.

### **4. Постановка краевой задачи (3 лабораторная работа)**

**Обезразмеривание переменных:**  
При расчете устойчивости значения шагов на квадратной сетке задаются следующим образом:  
$$t = p \tau, \; \; \; \; p = 1,..., P, \; \; \; \; t \in [0,1]$$
$$x = m h, \; \; \; \; m = 1,..., M, \; \; \; \; x \in [0,1]$$
Переменные $t$ и $x$ меняются от $0$ до $1$, и условия устойчивости применяются соответственно. Нам нужно изменить масштаб для переменных (обезразмерить переменные), чтобы он удовлетворял нашему условию:
$$t \in [0,T], \; \; \; \; \overline{t} = \frac{t}{T}, \; \; \; \; \overline{t} \in [0,1]$$
$$x \in [0,L], \; \; \; \; \overline{x} = \frac{x}{L}, \; \; \; \; \overline{x} \in [0,1]$$
Длина струны $L$ равна $0.1м$, время $T$ возьму равным $1сек$.

**Краевая задача:**  
Перепишу производные функции $U(\overline{x},\overline{t})$ по $t$ и $x$ согласно правилу взятия производной от сложной функции:
$$U_t=U_{\overline{t}} \cdot \overline{t}_{t}= \frac{1}{T} \cdot U_{\overline{t}}$$
$$U_{tt}= \frac{1}{T^2} \cdot U_{\overline{t} \overline{t}}$$
$$U_x=U_{\overline{x}} \cdot \overline{x}_{x}= \frac{1}{L} \cdot U_{\overline{x}}$$
$$U_{xx}= \frac{1}{L^2} \cdot U_{\overline{x} \overline{x}}$$

Тогда исходная краевая задача будет выглядеть так:  
$$\begin{cases} \frac{1}{T^2} \cdot U_{\overline{t} \overline{t}} = c^2 \cdot \frac{1}{L^2} \cdot U_{\overline{x} \overline{x}} + \frac{\alpha}{\rho'} \exp \left(-\frac{\left(\overline{x}- \frac{x_0}{L}\right)^{2}}{(\frac{\gamma}{L})^{2}} \right) \\ U \left( 0,\overline{x} \right) = 0 \\ \frac{1}{T} \cdot U_{\overline{t}} \left( 0,\overline{x} \right) = 0 \\ U \left( \overline{t},0 \right) = 0 \\ U \left( \overline{t},1 \right) = 0 \end{cases}$$

**Дополнительный вопрос:**  
Решение после одной секуды продолжает оставаться правильным благодаря свойству линейности уравнения и свойству сохранения формы волнового процесса. Колебания последовательны и периодичны, поэтому, даже если мы рассматриваем только один временной отрезок, физические свойства системы остаются неизменными, а волна будет продолжать распространяться, сохраняя свою форму.

Составлю матрицу для решения дифференциального уравнения методом прогонки:
$$A_m = -\frac{c^2}{L^2h^2}; \; \; B_m = -\frac{c^2}{L^2h^2}; \; \; C_m = -\left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{c^2}{L^2h^2} \right); \; \; \; \; (2.7)$$
$$F_m = - \left( \frac{\alpha}{\rho'} \exp \left(-\frac{\left(m h - \frac{x_0}{L}\right)^{2}}{(\frac{\gamma}{L})^{2}} \right) - \frac{1}{T^2 \tau^2} \cdot (U_m^{p-1} - 2U_m^p) \right)$$
Из граничных условий:
$$U_0^p = 0 \; \; \implies \; \; \chi_1 = \mu_1 = 0$$
$$U_M^p = 0 \; \; \implies \; \; \chi_2 = \mu_2 = 0$$
Тогда матрица будет иметь вид:
$$AU = F \; \; \; \; (2.8)$$
$$ A = \left[ \begin{matrix}
1 & 0 & ... & ... & ... & ... & ... \\
-\frac{c^2}{L^2h^2} & \left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{c^2}{L^2h^2} \right) & -\frac{c^2}{L^2h^2} & 0 & ... & ... & ... \\
0 & -\frac{c^2}{L^2h^2} & \left( \frac{1}{T^2 \tau^2} + 2 \cdot \frac{c^2}{L^2h^2} \right) & -\frac{c^2}{L^2h^2} & 0 & ... & ... \\

... & ... & ... & ... & ... & 0 & 1
\end{matrix} \right]$$
$$ U = \left[ \begin{matrix}

\end{matrix} \right]
\; \; \; \; F = \left[ \begin{matrix}

&\frac{\alpha}{\rho'} \exp \left(-\frac{\left(2 h - \frac{x_0}{L}\right)^{2}}{(\frac{\gamma}{L})^{2}} \right) - \frac{1}{T^2 \tau^2} \cdot (U_2^{p-1} - 2U_2^p) \\

&\frac{\alpha}{\rho'} \exp \left(-\frac{\left((M-1) h - \frac{x_0}{L}\right)^{2}}{(\frac{\gamma}{L})^{2}} \right) - \frac{1}{T^2 \tau^2} \cdot (U_{M-1}^{p-1} - 2U_{M-1}^p) \\

\end{matrix} \right]$$

### **6. Результат моделирования (5 лабораторная работа)**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation


#Решение с помощью явного шаблона
def solve_with_explicit_template(T, P, L, M, N, rho, alpha, x_0, gamma):
    U = np.zeros((M+1, P+1)) #Матрица для хранения значений функции в узлах
    #Параметры
    tau = 1/P #шаг по оси Ot (с)
    h = 1/M #шаг по оси Ox (м)
    c = np.sqrt(N/rho) #Коэффициент
    #Граничные условия
    for p in range(P+1):
        U[0, p] = 0
        U[M, p] = 0
    #Начальные условия
    for m in range(1, M):
        U[m, 0] = 0
        U[m, 1] = tau**2/2 * alpha/rho * np.exp(-(m*h-x_0/L)**2/(gamma/L)**2)
    #Нахождение значений U в остальных узлах
    for p in range(1, P):
        for m in range(1, M):
            U[m, p+1] = T**2*tau**2 * ( alpha/rho*np.exp(-(m*h-x_0/L)**2/(gamma/L)**2) + c**2/(L**2*h**2)*U[m-1, p] + 2*(1/(T**2*tau**2)-c**2/(L**2*h**2))*U[m, p] + c**2/(L**2*h**2)*U[m+1, p] - 1/(T**2*tau**2)*U[m, p-1] )
    return U


#Решение с помощью неявного шаблона
def solve_with_implicit_template(T, P, L, M, N, rho, alpha, x_0, gamma):
    U = np.zeros((M+1, P+1)) #Матрица для хранения значений функции в узлах
    #Параметры
    tau = 1/P #шаг по оси Ot (с)
    h = 1/M #шаг по оси Ox (м)
    c = np.sqrt(N/rho) #Коэффициент
    #Граничные условия
    for p in range(P+1):
        U[0, p] = 0
        U[M, p] = 0
    #Начальные условия
    for m in range(1, M):
        U[m, 0] = 0
        U[m, 1] = tau**2/2 * alpha/rho * np.exp(-(m*h-x_0/L)**2/(gamma/L)**2)
    
    #Коэффициенты матрицы
    A = -( c**2/(L**2*h**2) )
    B = -( c**2/(L**2*h**2) )
    C = -( 1/(T**2*tau**2) + 2*c**2/(L**2*h**2) )
    #Прогонка
    for p in range(1, P):
        #Коэффициенты для прогонки
        a = list(np.zeros(M+1))
        b = list(np.zeros(M+1))
        F = list(np.zeros(M+1))
        #Прямой ход
        for m in range(1, M):
            F[m] = -( alpha/rho*np.exp(-(m*h-x_0/L)**2/(gamma/L)**2) - 1/(T**2*tau**2) * (U[m, p-1] - 2*U[m, p]) )
        for m in range(1, M+1):
            a[m] = B/(C - a[m-1]*A)
        for m in range(1, M+1):
            b[m] = (A*b[m-1] + F[m-1])/(C - a[m-1]*A)
        #Обратный ход
        for m in np.arange(M-1, 0, -1):
            U[m, p+1] = a[m+1]*U[m+1, p+1] + b[m+1]
    return U


#Визуализация результатов
def draw_surface(P, M, U, title):
    x = np.linspace(0, 1, M+1)
    t = np.linspace(0, 1, P+1)
    xx, tt = np.meshgrid(x, t)
    uu = U.T
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(xx, tt, uu, cmap='viridis')
    ax.set_xlabel('x')
    ax.set_ylabel('t')
    ax.set_zlabel('U')
    ax.set_title(title)
    ax.view_init(elev=0, azim=0)
    plt.show()
```

```python
#Изменяемые параметры задачи
alpha = 100 #амплитуда силы (Н)
x_0 = 0.5 #точка приложения силы (м)
gamma = 0.12 #ширина распределения силы (м)


#Неизменяемые параметры задачи
T = 0.00055 #временной отрезок (c)
P = 20000 #Количество точек по оси Ot
L = 1 #длина струны (м)
M = 200 #Количество точек по оси Ox
N = 40 #Cила преднатяжения струны (Н)
rho = 0.016 #Плотность на единицу длины (кг/м)


U_e_t = solve_with_explicit_template(T, P, L, M, N, rho, alpha, x_0, gamma)
draw_surface(P, M, U_e_t, 'Колебания струны расчитанные с помощью явного шаблона')
```
