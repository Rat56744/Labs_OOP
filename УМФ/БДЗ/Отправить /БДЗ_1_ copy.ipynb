 **Задание 1.**   
Найти общее решение однородного линейного Д.У.Ч.П.
$$e^{x}U_x + \frac{3}{y} U_y = 0$$

 **Волшебная фраза:**    
Пред нами Д.У. в частных производных, линейное, однородное, первого порядка   
следовательно решение будем искать в виде:  
$$U(x,y)=w(g(x,y)),$$
где $w(\xi)$ - произвольная функция от одного переменного,  
$g(x,y)$ - единственный первый интеграл.  

 **Первый интеграл:**   

$$\frac{dx}{e^{x}} = \frac{y dy}{3}$$
$$\int \frac{1}{e^{x}} dx = \frac{1}{3}\int y dy$$
$$-e^{-x} = \frac{1}{6}y^{2} + C$$
$$g(x,y) = C = y^{2} + 6{e^{-x}}$$
 **Ответ:**   
$$U(x,y) = w \left( y^{2} + 6{e^{-x}} \right)$$

 **Задание 3.** Найти общее решение квазилинейного уравнения.  
$$\frac{1}{sin(x)}U_x+e^{-y}xU_y=U(y^{2}+1)$$

 **Волшебная фраза:**   
Пред нами квазилинейное, неоднородное ДУЧП первого порядка следовательно общее решение будем искать в виде:  
$$\psi(x,y,U)=w(g_1(x,y,U),g_2(x,y,U))=0,$$  
$$ g_2(x,y,U) = \widetilde{w}(g_1(x,y,U))$$ 
$$ U = g_2^{-1}(\widetilde{w}(g_1(x,y,U)))$$
где $w(\xi,\eta)$ - произвольная непрерывно диффиренцируемая функция от двух переменных,  
$g_1(x,y,U),g_2(x,y,U)$ - первые интегралы.  

 **Решение:**   
$$\sin(x)dx=e^{y} dy=\frac{dU}{U(y^{2}+1)}$$


 **Первый интеграл номер 1:**   

$$\sin(x)dx=e^{y} dy$$
$$-\cos(x) = e^{y} + C_1$$
$$g_1(x,y) = C_1 = e^{y} + \cos(x)$$


 **Первый интеграл номер 2:**   

$$e^{y}(y^{2}+1)dy=\frac{dU}{U}$$
Интегрируя по частям получим:
$$e^{y}(y^{2}-2y+3)dy=\ln\left|U\right|+C_2$$
$$g_2(y,U) = C_2 = e^{y}(y^{2}-2y+3)-\ln\left|U\right|$$
 **Функция $\psi$:**   
$$\psi(x,y,U)=w \left(e^{y} + \cos(x), \; \; e^{y}(y^{2}-2y+3)dy-\ln\left|U\right|\right)=0$$
 **$U(x,y)$:**  
$$e^{y}(y^{2}-2y+3)-\ln\left|U\right| = \widetilde{w} \left( e^{y} + \cos(x) \right)$$
$$\ln\left|U\right| = e^{y}(y^{2}-2y+3) + \widetilde{w} \left( e^{y} + \cos(x) \right) $$
 **Ответ:**  
$$U(x,y) =e^{\left( e^{y}(y^{2}-2y+3) + \widetilde{w} \left( e^{y} + \cos(x) \right) \right)}  $$

 **Задание 4.** Найти общее решение квазилинейного уравнения.  
$$(x^{2}+9)U_x+yU_y=xU$$

 **Волшебная фраза:**   
Пред нами квазилинейное, неоднородное ДУЧП первого порядка следовательно общее решение будем искать в виде:  
$$\psi(x,y,U)=w(g_1(x,y,U),g_2(x,y,U))=0,$$  
$$ g_2(x,y,U) = \widetilde{w}(g_1(x,y,U))$$ 
$$ U = g_2^{-1}(\widetilde{w}(g_1(x,y,U)))$$
где $w(\xi,\eta)$ - произвольная непрерывно диффиренцируемая функция от двух переменных,  
$g_1(x,y,U),g_2(x,y,U)$ - первые интегралы.  

 **Решение:**   
$$\frac{dx}{x^{2}+9}=\frac{dy}{y}=\frac{dU}{xU}$$


 **Первый интеграл номер 1:**   

$$\frac{dx}{x^{2}+9}=\frac{dy}{y}$$
$$\int\frac{dx}{x^{2}+9}=\int\frac{dy}{y}$$
$$\frac{1}{3} \arctan(\frac{x}{3}) = \ln\left|y\right|+C_1 $$
$$g_1(x,y) = C_1 = \frac{1}{3} \arctan(\frac{x}{3})-\ln\left|y\right|$$


 **Первый интеграл номер 2:**   

$$\frac{xdx}{x^{2}+9}=\frac{dU}{U}$$
$$\int\frac{xdx}{x^{2}+9}=\int\frac{dU}{U}$$
$$\frac{1}{2}\ln(x^{2}+9) = \ln\left|U\right|+C_2$$
$$g_2(x,U) = C_1 = \frac{1}{2}\ln(x^{2}+9)-\ln\left|U\right|$$


 **Функция $\psi$:**   
$$\psi(x,y,U)=w \left(\frac{1}{3}\arctan(\frac{x}{3})-\ln\left|y\right|, \; \; \frac{1}{2}\ln(x^{2}+9)-\ln\left|U\right|\right)=0$$


 **$U(x,y)$:**  
$$\frac{1}{2}\ln(x^{2}+9)-\ln\left|U\right| = \widetilde{w} \left( \frac{1}{3} \arctan(\frac{x}{3})-\ln\left|y\right|\right)$$
$$\ln\left|U\right| = \frac{1}{2}\ln(x^{2}+9) - \widetilde{w} \left( \frac{1}{3} \arctan(\frac{x}{3})-\ln\left|y\right|\right) $$


 **Ответ:**  
$$U(x,y) =e^{\frac{1}{2}\ln(x^{2}+9) - \widetilde{w} \left( \frac{1}{3}\arctan(\frac{x}{3})-\ln\left|y\right|\right)}  $$
