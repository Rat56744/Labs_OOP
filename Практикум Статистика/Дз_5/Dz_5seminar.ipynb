```python
import numpy as np
import scipy.stats as sts
import scipy.special as sc
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.stats.weightstats
import statsmodels.stats.proportion
%matplotlib inline
```

# Задание 1

```python
def gav(n):
    m = 0
    Sigma = 1 # Параметры нормального распределения
    betta = 0.95
    N = 100 # Количество выборок
    count = 0 # Счетчик плохих интервалов
    for i in range(-1,N-1):
        X = np.random.normal(m, Sigma, n)
        m_v = np.mean(X)
        S = np.var(X, ddof=1)
        t = sts.t(n-1).ppf((1 + betta)/2)
        m_min = m_v - t * S /np.sqrt(n)
        m_max = m_v + t * S / np.sqrt(n)

        if(m_min > 0 or m_max < 0): # Проверяем сколько интервалов не содержит m
            count+=1
        plt.plot([i + 2, i + 2], [m_min, m_max], color='orange') # Строим наши отрезки

    plt.axhline(m, color= '#6f0035')
    print('Количество интервалов в которые вошло реальное мат ожидание:', 100 - count)
    plt.xlim(-1, 105)
    plt.ylim(-0.4, 0.4)

    plt.grid(True)

gav(200)
```

### Мы видим, что количество интервалов в которые вошло реальное мат ожидание примерно равно коэфиценту бетта умноженому на 100. Т.е. наш доверительный интервал работает именно так, как мы и предолагали

```python
gav(500)
```
