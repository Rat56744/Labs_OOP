```python
x_n = np.linspace(-4 , 4 ,100)
y_n = np.linspace(-4 , 4 ,99)

F_x_n = 1/2 + (1/np.pi)*np.arcsin(x_n/ 4)
f_x_n = np.diff(F_x_n)

plt.plot(y_n , f_x_n, lw = 3)
plt.title('График плотности распределения $f(x)$')
plt.grid(True)
```

Пункт (2)

```python
print("Закон арксинуса моды не имеет")


exp = '1/2 + asin(x/4)/pi'
ans = sp.solve(exp, x)
print("Медиана равна: ", ans)

def F(x):
    return x/(4*np.pi*np.sqrt(1 - x**2/16))
result = integrate.quad(F , -4, 4)
print("Математическое ожидание: " , result )

def F2(x):
    return x**2/(4*np.pi*np.sqrt(1 - x**2/16))
result2 = integrate.quad(F2 , -4, 4)
print("Дисперсия равна:", result2)

exp = '1/2 + asin(x/4)/pi - 0.25'
qw1 = sp.solve(exp, x)
print("Квантиль порядка 0,25 равен: ", qw1)

exp = '1/2 + asin(x/4)/pi - 0.75'
qw2 = sp.solve(exp, x)
print("Квантиль порядка 0,75 равен: ", qw2)

def F101(x):
    return x**2/(4*np.pi*np.sqrt(1 - x**2/16))
result2 = integrate.quad(F101 , -4, 4)
print("Начальный момент второго порядка:", result2)

def F103(x):
    return x**3/(4*np.pi*np.sqrt(1 - x**2/16))
resultx = integrate.quad(F103 , -4, 4)
print("Начальный момент третьего порядка:", resultx)
```

Пункт (1)


```python
raspred1 = sts.chi2(2)
raspred2 = sts.chi2(6)
raspred3 = sts.chi2(8)
raspred4 = sts.chi2(10)

x_n = np.linspace(0 , 10 , 100)

f_x_n1 = raspred1.pdf(x_n)
f_x_n2 = raspred2.pdf(x_n)
f_x_n3 = raspred3.pdf(x_n)
f_x_n4 = raspred4.pdf(x_n)

plt.plot(x_n , f_x_n1, label = '2 param',lw = 3)
plt.plot(x_n , f_x_n2, label = '6 param',lw = 3)
plt.plot(x_n , f_x_n3, label = '8 param',lw = 3)
plt.plot(x_n , f_x_n4, label = '10 param',lw = 3)

plt.legend(title ='Parameters')

plt.title('Графики плотности распределения $f(x)$')
plt.grid(True)
```

Пункт (2)

```python
raspred1 = sts.chi2(2)
raspred2 = sts.chi2(6)
raspred3 = sts.chi2(8)
raspred4 = sts.chi2(10)

x_n = np.linspace(0 , 10 , 100)

f_x_n1 = raspred1.cdf(x_n)
f_x_n2 = raspred2.cdf(x_n)
f_x_n3 = raspred3.cdf(x_n)
f_x_n4 = raspred4.cdf(x_n)

plt.plot(x_n , f_x_n1, label = '2 param',lw = 3)
plt.plot(x_n , f_x_n2, label = '6 param',lw = 3)
plt.plot(x_n , f_x_n3, label = '8 param',lw = 3)
plt.plot(x_n , f_x_n4, label = '10 param',lw = 3)

plt.legend(title ='Parameters')

plt.title('Графики Функций распределения $F(x)$')
plt.grid(True)
```

```python
raspred1 = sts.chi2(2)
raspred2 = sts.chi2(6)
raspred3 = sts.chi2(8)
raspred4 = sts.chi2(10)

x_n = np.linspace(0 , 10 , 100 )
y_n = np.linspace(0 ,1 , 100)

f_x_n1 = raspred1.cdf(x_n) 
f_x_n2 = raspred2.cdf(x_n) 
f_x_n3 = raspred3.cdf(x_n)
f_x_n4 = raspred4.cdf(x_n) 

plt.plot(x_n , f_x_n1, label = '2 param',lw = 3)
plt.plot(x_n , f_x_n2, label = '6 param',lw = 3)
plt.plot(x_n , f_x_n3, label = '8 param',lw = 3)
plt.plot(x_n , f_x_n4, label = '10 param',lw = 3)

plt.legend(title ='Parameters')
plt.ylabel('Порядок')
plt.xlabel('Квантили')

plt.title('Графики Зависимости порядка от квантилей ')
plt.grid(True)
```
