# Задание 4 #
1) Преобразуйте данные, применив к выборкам А, В и С преобразование Бокса-Кокса (далее АА, ВВ и СС - преобразованные данные).

2) Постройте гистограммы А и AA, B и BB, С и СС. Сопоставьте
гистограммы прообразов и образов. Результаты прокомментируйте.

3) С помощью критерия Шапиро - Уилка (используйте функцию
shapiro () модуля scipy.stats) проверьте для всех шести выборок гипотезы о том, что выборки принадлежат нормально распределенным генеральным совокупностям. Результаты сопоставьте и прокомментируйте.

4) Для всех шести выборок постройте Q-Q - график. Сопоставьте
графики прообразов и образов. Результаты прокомментируйте и сопоставьте с результатами п. 3).

```python
V = pd. read_excel("Data_7_1.xlsx", names = ['Aimport numpy as np
import matplotlib.pyplot as plt
', 'B', 'C' , 'D' , 'E' ])
selections= ['A', 'B', 'C', 'D', 'E' ]
V = V.drop(index = 0)
V = V[V != '-' ]
V = V.astype(float)

# Обработка столбца A

A = V['A'].dropna().values.tolist()
A_uncide = [a for a in A if a > 0]  # Удаляем нолики(чисел меньше нуля у нас нет)
A_modificated = sts.boxcox(A_uncide)  # Преобразование Бокса-Кокса
A_modificated = A_modificated[0]  # Извлечение преобразованных данных

# Обработка столбца B

B = V['B'].dropna().values.tolist()
B_uncide = [b for b in B if b > 0]  # Удаляем нолики(чисел меньше нуля у нас нет)
B_modificated = sts.boxcox(B_uncide)  # Преобразование Бокса-Кокса
B_modificated = B_modificated[0]  # Извлечение преобразованных данных

# Обработка столбца C

C = V['C'].dropna().values.tolist()
C_uncide = [c for c in C if c > 0]  # Удаляем нолики(чисел меньше нуля у нас нет)
C_modificated = sts.boxcox(C_uncide)  # Преобразование Бокса-Кокса
C_modificated = C_modificated[0]  # Извлечение преобразованных данных

plt.figure(figsize=[6,4])
plt.hist(A, color='#44944A',edgecolor='#000000', density=True)
plt.title('Гистограмма А')
plt.grid()
plt.figure(figsize=[6,4])
plt.hist(A_modificated, color='#44944A',edgecolor='#000000', density=True)
plt.title('Гистограмма AA ' )
plt.grid()

plt.figure(figsize=[6,4])
plt.hist(B, color='#FFDC33',edgecolor='#000000', density=True)
plt.title('Гистограмма B')
plt.grid()
plt.figure(figsize=[6,4])
plt.hist(B_modificated, color='#FFDC33',edgecolor='#000000', density=True)
plt.title('Гистограмма BB ' )
plt.grid()

plt.figure(figsize=[6,4])
plt.hist(C, color='#9B2D30',edgecolor='#000000', density=True)
plt.title('Гистограмма C')
plt.grid()
plt.figure(figsize=[6,4])
plt.hist(C_modificated, color='#9B2D30',edgecolor='#000000', density=True)
plt.title('Гистограмма CC ' )
plt.grid()
```

```python
print(f"p значение для A  {sts.shapiro(A) [1]}")
print(f"p значение для AA {sts.shapiro(A_modificated) [1]}")
print(f"р значение для B  {sts.shapiro(B) [1]}")
print(f"p значение для BB {sts.shapiro(B_modificated) [1]}")
print(f"р значение для C  {sts.shapiro(C) [1]}")
print(f"p значение для CC {sts.shapiro(C_modificated) [1]}")

    #   A и AA

sts.probplot(A, plot=plt)
plt.grid(True)
plt.title("A")
plt.figure()
sts.probplot(A_modificated, plot=plt)
plt.grid(True)
plt.title("AA")

    #   B и BB

plt.figure()
sts.probplot(B, plot=plt)
plt.grid(True)
plt.title("B")
plt.figure()
sts.probplot(B_modificated, plot=plt)
plt.grid(True)
plt.title("BB")
plt.figure()

    #   C и CC

sts.probplot(C, plot=plt)
plt.grid(True)
plt.title("C")
plt.figure()
sts.probplot(C_modificated, plot=plt)
plt.grid(True)
plt.title("CC")
plt.figure()
```
