```python
n1 = 40
lambd = 1

y_n = np.random.uniform(0,lambd,n1)
x1_n = (-1/lambd)*np.log(y_n)

xx = np.linspace(-15,15,n1)
F_x = sts.expon(lambd).cdf(xx)
ecdf = ECDF (x1_n)

plt.step(ecdf.x , ecdf.y , label = 'Эмпирическая функция распределения')
plt.plot(xx , F_x , lw=3 , color = 'r' , label = 'Теоретическая функция распределения для n = 40')


plt.legend(loc = 'lower left')
plt.grid(True)
```
