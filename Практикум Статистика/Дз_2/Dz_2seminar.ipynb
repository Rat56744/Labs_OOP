## Задаем выборку

```python
a = 8
x1 = 100
x2 = 1000
p1 = [0.1, 0.125, 0.07, 0.16, 0.18, 0.05, 0.15, 0.165]
p2 = [0.123, 0.134, 0.064, 0.153, 0.168, 0.045, 0.147, 0.166]
x1n = np.random.choice(a, x1, True, p1)
x2n = np.random.choice(a, x2, True, p2)

print("Выборка номер один с параметрами size = 100 и a = 8:", '\n',x1n, '\n')
print("Выборка номер два с параметрами size = 1000 и a = 8:", '\n',x2n)
```

## Первая выборка

```python
print('Выборка:' , '\n', x1n)

z = np.sort(list(set(x1n)))
print('\n','Набор различных значений выборки:', '\n', 'zk =', z)

pz = np.bincount(x1n)/x1
print('\n','Относительные частоты различных элементов:', '\n', pz)

z_pz = np.stack((z, pz))
print('\n','Статистический ряд:', '\n', 'zk \t pk')
for k in range(len(z)):
    print(z[k], '\t', pz[k])
plt.title('Визуализация статистического ряда номер 1')
plt.bar(z, pz, width = 0.5)
```

## Задание 2

## Аналитический закон

```python
p = 0.6
q = 1 - p
K_val = []
probabilities = []
for i in range(1,7):
    if i == 6:
        probability = p ** i + p**(i-1)*q
    else:
        probability = (p ** (i - 1)) * q
    K_val.append(i)
    probabilities.append(probability)
print('Количество приборов','\t','Вероятность')
for s in range(0,6):
    print(K_val[s],'\t','\t','\t', f"{probabilities[s]:.6f}")       
```

## Распределение:

```python
plt.bar(K_val, probabilities, width=0.3, color = '#B00000')
plt.title('Закон распределения случайной величины')
plt.grid(True)
```

## Находим Математическое ожидание:

```python
M = 0
for i in range(0,6):
    M = M + K_val[i] * probabilities[i]
print("Математическое ожидание равно:", f"{M:.4f}")
```

## Находим Дисперсию:

```python
D = 0
for i in range(0,6):
    D = D + K_val[i]**2 * probabilities[i]
D = D - M**2
print("Дисперсия равна:", f"{D:.4f}")
```

## Компьютерное моделирование:

```python
pa = [0.4, 0.6]
am = [0, 1]
n = 300
x = []
for i in range(1, 301):
    for j in range(1,7):
        if (np.random.choice(am, 1, True, pa) == 0):
            x.append(j)
            break
print('Выборка из количества испытанных элементов размером 300: ',x)
z = np.sort(list(set(x)))
pzi = [0]*7
pz = np.bincount(x)[1:]/n
for i in range(0, 6):
    pzi[i] = float(pz[i])
pz_norm = []
for i in range(0, 6):
    pz_norm.append(round(pzi[i], 3))
print('Относительные частоты: ','\n',pz_norm)

print('Статистический ряд:','\n','zk\t pk')
for k in range(0,len(z)):
    print(z[k], '\t', f"{pz_norm[k]:.3f}")       
```

## Теорема Чебышева

```python
n = 300
M = []
k = np.linspace(1, n, n-1)
for i in range(1, n):
    pa = [0.4, 0.6]
    am = [0, 1]
    x = []
    for j in range(i):
        kol_prib = 1
        R = np.random.choice(am, 1, pa)
        while((R[0]==1) & (kol_prib < 6)):
            R = np.random.choice(am, 1, pa)
            kol_prib +=1
        x.append(kol_prib)
    M.append(sum(x)/i)
print(x,'\n', M)
```

```python
plt.scatter(k,M, color = '#1A153F')
plt.plot([0, n], [sum(M)/n, sum(M)/n], color = '#F64A46')
plt.grid()
plt.title("Теорема Чебышева")
plt.xlabel("n")
plt.ylabel("X")
plt.legend(['Среднее арифметическое значение СВ','Теоретическое матожидание'], loc = 8)
```
