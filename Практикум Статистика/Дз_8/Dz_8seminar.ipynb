## Задание 2.
Сгенерируйте две выборки $(А и В)$ различных объемов из нормально распределенныхгенеральных совокупностей с параметрами $m_A, \sigma_A$ и $m_B, \sigma_B$ (возьмитезначения $m_A$ и $m_B$ близкими, но не
равными, а $\sigma_A$ и $\sigma_B$ выберите такими, чтобы близки (но не равны)
были их квадраты). Проверьте гипотезу о равенстве дисперсий по критерию Фишера (см. п. 2 подраздела 8.2) на уровне значимости 0,05.

Сформулируйте выводы.
Если гипотеза подтвердится, то сгенерируйте две другие выборки
(С и D), для которых гипотеза отклонится (при необходимости поварьируйте параметры, поэкспериментируйте, чтобы добиться заданного
результата).
Если гипотеза о равенстве дисперсий выборок А и В будет отклонена, то сгенерируйте две другие выборки (С' и D'), для которых
гипотеза подтвердится (при необходимости поварьируйте параметры,
поэкспериментируйте, чтобы добиться заданного результата).

```python
def ДОнбаСС(m_a, sigm_a, m_b, sigm_b):
    alpha = 0.05
    # Размеры выборок
    N_a = 3000
    N_b = 7000
    # Распределения
    A = np.random.normal(m_a, sigm_a, N_a)
    B = np.random.normal(m_b, sigm_b, N_b)
    # Оценки дисперсий
    S_a = np.var(A , ddof = 1)
    S_b = np.var(B , ddof = 1)
    # Статистика(Критерий Фишера)
    F = S_a/S_b
    # Критические значения
    D_alpha_low = sts.f(N_a - 1, N_b - 1).ppf(alpha/2)
    D_alpha_up = sts.f(N_a - 1, N_b - 1).ppf(1 - alpha/2)

    print('S_a = ', S_a)
    print('S_b = ', S_b)
    print('Значение статистики F = ', F)
    print('Критическая область: (-∞, ', D_alpha_low,') ∪ (', D_alpha_up,',+∞)')
    # Определяем принять или овергнуть гипотезу
    if(D_alpha_low < F < D_alpha_up):
        print('Гипотеза принимается\n')
    else:
        print('Гипотеза отклоняется\n')

ДОнбаСС(1, 2 , 1.1 , 2.2)
ДОнбаСС(1, 2 , 1.01 , 2.005)

print(sts.norm(0, 1).ppf(1 - 0.05/2))
```

# Задние 3.

Для выполнения задания используйте пару выборок,
смоделированных при выполнении задания 2, для которой гипотеза о
равенстве дисперсий подтвердилась. Проверьте гипотезу о равенстве
математических ожиданий. Укажите диапазон значений $\alpha$, при которых гипотеза принимается.

```python
# Размеры выборок
N_a = 3000
N_b = 7000

# Параметры распределения
m_a = 1 
sigm_a = 2
m_b = 1.01
sigm_b = 2.005

# Распределения
A = np.random.normal(m_a, sigm_a, N_a)
B = np.random.normal(m_b, sigm_b, N_b)

# Проверка гипотезы о равенстве математических ожиданий
st = sts.ttest_ind(A, B, equal_var=True, axis=0) [0]
p = sts.ttest_ind(A, B, equal_var=True, axis=0) [1]
print ('Выборочное значение статистики =', st)
print ('p =', p)
print ('При альфа <= ', p,'гипотеза принимается')
```
