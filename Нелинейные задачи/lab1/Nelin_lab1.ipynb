```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)
F1 = np.cos(X)*np.cosh(Y) - X - 1
F2 = -np.sin(X)*np.sinh(Y) - Y
plt.contour(X, Y, F1, cmap = 'autumn')
plt.contour(X, Y, F2)
plt.grid(True)
```

```python
dx, dy = sp.symbols('x y')
orF1 = sp.cos(dx)*sp.cosh(dy) - dx - 1
orF2 = -sp.sin(dx)*sp.sinh(dy) - dy
F1x = sp.diff(orF1, dx)
F1y = sp.diff(orF1, dy)
F2x = sp.diff(orF2, dx)
F2y = sp.diff(orF2, dy)
print(F1x.subs({dx: 0, dy: 0}))
print(F1x, '\n', F1y)
print(type(F1x))
```

```python
xi = -2.3
yi = 1.6
d = 0.01
dx, dy = sp.symbols('x y')
orF1 = sp.cos(dx) * sp.cosh(dy) - dx - 1
orF2 = -sp.sin(dx) * sp.sinh(dy) - dy
F1x = sp.diff(orF1, dx)
F1y = sp.diff(orF1, dy)
F2x = sp.diff(orF2, dx)
F2y = sp.diff(orF2, dy)
while d < max(orF1.subs({dx: xi, dy: yi}), orF2.subs({dx: xi, dy: yi})):
    matr1 = sp.Matrix([[xi], [yi]])
    matrdf = sp.Matrix([[F1x.subs({dx: xi, dy: yi}), F1y.subs({dx: xi, dy: yi})],
                        [F2x.subs({dx: xi, dy: yi}), F2y.subs({dx: xi, dy: yi})]])
    matr3 = sp.Matrix([[orF1.subs({dx: xi, dy: yi})], 
                        [orF2.subs({dx: xi, dy: yi})]])
    
    invmatrdf = matrdf.inv()
    
    main = matr1 - np.dot(invmatrdf, matr3)
    
    result = main.evalf()  # Используем evalf для численного вычисления
    xi = result[0]
    yi = result[1]
print(xi, yi)
```
