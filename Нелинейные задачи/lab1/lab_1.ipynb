```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
#   
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

#   Z
F_1_x_y = np.cos(X)*np.cosh(Y) - X - 1
F_2_x_y = -np.sin(X)*np.sinh(Y) - Y

#   
plt.contour(X, Y, F_1_x_y, cmap='viridis')
plt.contour(X, Y, F_2_x_y, cmap='viridis')
plt.colorbar()  #   
plt.title('   Z = sin(sqrt(X^2 + Y^2))')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
