```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
# Создаем сетку координат
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Определяем функцию Z
F_1_x_y = np.cos(X)*np.cosh(Y) - X - 1
F_2_x_y = -np.sin(X)*np.sinh(Y) - Y

# Создаем контурный график
plt.contour(X, Y, F_1_x_y, cmap='viridis')
plt.contour(X, Y, F_2_x_y, cmap='viridis')
plt.colorbar()  # Добавляем цветовую шкалу
plt.title('Контурный график функции Z = sin(sqrt(X^2 + Y^2))')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
